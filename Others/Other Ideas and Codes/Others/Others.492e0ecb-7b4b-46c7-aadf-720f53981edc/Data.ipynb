# London Hydro data

```python
import os
os.getcwd()

csvs = [file for file in os.listdir('./Data/Muhammad/') if file.endswith('.csv')]
```

```python
import pandas as pd

# Read the weather data
weather = pd.read_csv('./Data/Muhammad/weather.csv')
weather = weather[['Datetime', 'Temperature', 'Humidity', 'WindSpeed']]

for csv in csvs:
    # Read the PJME data
    df = pd.read_csv(f'./Data/Muhammad/{csv}')
    
    # Merge the two DataFrames on the 'Datetime' column
    df = pd.merge(weather, df, on='Datetime')
    
    # Rename the 'PJME_MW' column to 'energy_consumption'
    df.rename(columns={'Datetime':'timestamp', 'PJME_MW': 'energy_consumption', 'Temperature':'temperature', 'Humidity':'humidity',  'WindSpeed':'wind_speed'}, inplace=True)

    df.to_csv(f'./Data/LondonHydro/{csv}')
    print(csv)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some test data
y_test = np.linspace(0, 10, 100)
y_pred = y_test + np.random.normal(0, 1, 100)  # Predicted data with some noise

# Plot with the provided custom colors
plt.figure(figsize=(10, 6))
plt.plot(y_test, label='True', color='#ea9595')  # Custom red color
plt.plot(y_pred, label='Predicted', color='#1f77b4')  # Custom blue color
plt.legend()
plt.title('Test Plot with Custom Colors')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.show()
```
