```python
import tensorflow as tf
import psutil
import GPUtil
from tabulate import tabulate
import gc

# Clear TensorFlow memory and free up GPU memory
def clear_gpu_memory():
    # Clear the TensorFlow backend session
    tf.keras.backend.clear_session()
    
    # Run garbage collection
    gc.collect()

    # Reset the memory growth setting to clear memory
    gpus = tf.config.list_physical_devices('GPU')
    if gpus:
        try:
            for gpu in gpus:
                tf.config.experimental.set_memory_growth(gpu, False)
                # Re-enable memory growth if needed
                tf.config.experimental.set_memory_growth(gpu, True)
        except RuntimeError as e:
            print(f"Could not reset GPU memory: {e}")

clear_gpu_memory()

# Proceed with your code if needed


# Check if TensorFlow is using a GPU
gpu_available = tf.config.list_physical_devices('GPU')

if gpu_available:
    print("GPU is available.")
else:
    print("Using CPU.")

# Get CPU usage
cpu_usage = psutil.cpu_percent(interval=1)

# Get RAM usage
memory = psutil.virtual_memory()
ram_usage = memory.percent
ram_free = memory.available / (1024 ** 3)  # Convert to GB
ram_total = memory.total / (1024 ** 3)     # Convert to GB

# Get GPU usage (if GPU is available)
gpu_info = []
if gpu_available:
    gpus = GPUtil.getGPUs()
    for gpu in gpus:
        gpu_info.append((
            gpu.id,
            gpu.name,
            f"{gpu.load * 100:.0f}%",
            f"{gpu.memoryFree / 1024:.2f} GB",
            f"{gpu.memoryUsed / 1024:.2f} GB",
            f"{gpu.memoryTotal / 1024:.2f} GB",
            f"{gpu.temperature} °C"
        ))

# Print CPU, RAM, and GPU usage
print(f"\nCPU Usage: {cpu_usage}%")
print(f"RAM Usage: {ram_usage}% ({ram_free:.2f} GB free out of {ram_total:.2f} GB)")

if gpu_info:
    print("\nGPU Usage:")
    print(tabulate(gpu_info, headers=("ID", "Name", "Load", "Free Memory", "Used Memory", "Total Memory", "Temperature")))
else:
    print("\nNo GPU detected or GPU information not available.")
```
